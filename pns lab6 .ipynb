```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Parameters for the normal distribution
mean = 50
std_dev = 5
num_samples = 1000
# Generate data points from a normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='skyblue', bins=30, stat='density', linewidth=1)
# Adding labels and title
plt.title('Histogram and Probability Density Function (PDF)\nfor Normal Distribution', fontsize=14)
plt.xlabel('Data Points', fontsize=12)
plt.ylabel('Density', fontsize=12)
# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
mean = 14
std_dev = 2.5
sample_size = 1000

# X range for plotting
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)

# I. Probability between 2.5 and 14
z1, z2 = (2.5 - mean) / std_dev, (14 - mean) / std_dev  # Z-scores
p1 = norm.cdf(z2) - norm.cdf(z1)
students_between_2_5_and_14 = int(p1 * sample_size)

# II. Probability below 12
z3 = (12 - mean) / std_dev
p2 = norm.cdf(z3)
students_below_12 = int(p2 * sample_size)

# III. Probability above 19
z4 = (19 - mean) / std_dev
p3 = 1 - norm.cdf(z4)
students_above_19 = int(p3 * sample_size)

# Print Results
print(f"I. Students between 2.5 and 14: {students_between_2_5_and_14}")
print(f"II. Students below 12: {students_below_12}")
print(f"III. Students above 19: {students_above_19}")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, label='Normal Distribution', color='black')

# Fill between 2.5 and 14
x_fill_1 = np.linspace(2.5, 14, 1000)
plt.fill_between(x_fill_1, norm.pdf(x_fill_1, mean, std_dev), color='skyblue', alpha=0.5, label='Between 2.5 and 14')

# Fill below 12
x_fill_2 = np.linspace(mean - 4*std_dev, 12, 1000)
plt.fill_between(x_fill_2, norm.pdf(x_fill_2, mean, std_dev), color='orange', alpha=0.5, label='Below 12')

# Fill above 19
x_fill_3 = np.linspace(19, mean + 4*std_dev, 1000)
plt.fill_between(x_fill_3, norm.pdf(x_fill_3, mean, std_dev), color='green', alpha=0.5, label='Above 19')

# Plot formatting
plt.title('Normal Distribution Curve with Marked Probabilities')
plt.xlabel('Test Scores')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
